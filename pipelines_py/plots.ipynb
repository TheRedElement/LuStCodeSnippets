```python
import matplotlib.pyplot as plt
import numpy as np
from LuStCodeSnippets_py.Styles import PlotStyles

palette, ls, markers, cmap, hatches = PlotStyles.tre_dark()
hatches = hatches[0:-1:3]
plt.rcParams["savefig.transparent"] = False
```

## Histograms

```python
X = np.array(np.random.randn(1000,3)) + np.array([-3,0,3])

width = 0.5
bins = np.arange(np.floor(X.min()), np.ceil(X.max()),width).round(2)

counts = np.array([np.histogram(xi, bins)[0] for xi in X.T])

fig, axs = plt.subplots(1)
for idx, c in enumerate(counts):

    axs.bar(bins[0:-1]+width/2, c, width=width, edgecolor=palette[idx], hatch=hatches[idx], alpha=0.5)
    axs.bar(bins[0:-1]+width/2, c, width=width, edgecolor=palette[idx], facecolor="none")   #add storng outline


plt.show()
```
